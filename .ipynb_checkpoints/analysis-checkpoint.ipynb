## Part 1: Exploratory Data Analysis(EDA)

```python
import geopandas as gpd

# Adjust the file path as needed
# data_path = 'LF2023_PRVI_240_IA/LF2023_LDist_240_PRVI/Spatial_Metadata/z90_prvi_0k.shp'
# data_path = 'LF2016_PRVI_200_IA/LF2016_Dist_200_PRVI/Spatial_Metadata/z90_prvi_0k.shp'
# data_path = 'LF2016_PRVI_200_IA/LF2016_EVC_200_PRVI/Spatial_Metadata/IAz90_SPATPLYGN.shp'
data_path = 'LF2016_PRVI_200_IA/LF2016_NVC_200_PRVI/Spatial_Metadata/IAz90_SPATPLYGN.shp'



gdf = gpd.read_file(data_path)
# Remove features with missing geometry
gdf = gdf[gdf.geometry.notnull()]
# Calculate the area of each feature
gdf['area'] = gdf.geometry.area
```

```python
import matplotlib.pyplot as plt

gdf.plot(column='area')
plt.show()
```

```python
import folium

latitude, longitude = 18.200178, -66.664513

m = folium.Map(location=[latitude, longitude], zoom_start=12)
folium.GeoJson(gdf).add_to(m)
m.save('map.html')
```

## Landfire dataset

```python
import landfire

# Obtain required layers for FlamMap landscape file
lf = landfire.Landfire(bbox="-107.70894965 46.56799094 -106.02718124 47.34869094")
lf.request_data(layers=["ELEV2020",   # elevation
                        "SLPD2020",   # slope degrees
                        "ASP2020",    # aspect
                        "220F40_22",  # fuel models
                        "220CC_22",   # canopy cover
                        "220CH_22",   # canopy height
                        "220CBH_22",  # canopy base height
                        "220CBD_22"], # canopy bulk density
                output_path="./test_flammap.zip")
```

## TIF Analysis

```python
import rasterio
from rasterio.plot import show

# Path to your LANDFIRE GeoTIFF file
# lf_file_path = 'LF2016_PRVI_200_IA\LF2020_FBFM13_200_PRVI\Tif\LV20_F13_200.tif'
lf_file_path = 'test_flammap\j133d17fe0a994bbc9f9439a5560d5a57.tif'



# Open the file
with rasterio.open(lf_file_path) as src:
    lf_data = src.read(1)  # Read the first band into a 2D array
    show(lf_data)  # Visualize the data
```

```python
import numpy as np

# Mask out no-data values, assuming they are specified in the dataset
lf_data_masked = np.ma.masked_equal(lf_data, src.nodata)

# Compute statistics
mean_val = lf_data_masked.mean()
median_val = np.ma.median(lf_data_masked)
std_dev = lf_data_masked.std()

print(f"Mean: {mean_val}, Median: {median_val}, Standard Deviation: {std_dev}")
```

```python
import matplotlib.pyplot as plt

# Example: Visualize the data with a color map
plt.imshow(lf_data_masked, cmap='viridis')
plt.colorbar()
plt.title('LANDFIRE 2016 Remap Visualization')
plt.show()

# Save output to a new GeoTIFF (if needed)
output_path = 'output_file.tif'
with rasterio.open(output_path, 'w', **src.meta) as dst:
    dst.write(lf_data_masked.filled(src.nodata), 1)
```
